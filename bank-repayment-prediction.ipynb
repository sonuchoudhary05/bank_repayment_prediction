```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv("../input/loan-repayment-prediction/loan_data_final.csv")
df.head()
```

```python
x =df.iloc[:,1:14]
x
```

```python
y =df.iloc[:,0]
y
```

```python
purpose1 =pd.get_dummies(df['purpose'],drop_first=True)
purpose1
```

```python
x = x.drop('purpose',axis=1)
x
```

```python
x =pd.concat([x,purpose1],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
model =DecisionTreeClassifier(criterion ="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
model.fit(x_train,y_train)
```

```python
y_pred =model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
score =accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```
